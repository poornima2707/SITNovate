```python
import pandas as pd
import numpy as np

# Function to remove only extreme outliers
def remove_extreme_outliers(df):
    df_cleaned = df.copy()

    def iqr_filter(group):
        Q1 = group["Sales"].quantile(0.25)
        Q3 = group["Sales"].quantile(0.75)
        IQR = Q3 - Q1
        upper_bound = Q3 + 3.5 * IQR  #  Increased threshold to remove only extreme spikes
        return group[group["Sales"] <= upper_bound]  # Keep all valid values

    df_cleaned = df_cleaned.groupby(["Year", "Month"], group_keys=False).apply(iqr_filter)

    return df_cleaned

# Apply outlier removal
df_cleaned = remove_extreme_outliers(df)

# Check removed values
outliers_removed = df[~df.index.isin(df_cleaned.index)]
print("Total extreme outliers removed:", len(outliers_removed))
print(outliers_removed[["Year", "Month", "Sales"]])  # Show extreme outliers removed
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
import warnings

warnings.filterwarnings("ignore")

#  1 Load Data (Modify path based on your dataset)
df = pd.read_csv("FINAL_DATASET.csv", parse_dates=["Order Date"], dayfirst=True)  # Ensure DD-MM-YYYY format
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")  # Explicit conversion

# Drop any invalid date entries
df = df.dropna(subset=["Order Date"])

#  2 Aggregate Sales by Month
df["Month"] = df["Order Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()
monthly_sales["Month"] = monthly_sales["Month"].astype(str)  # Convert period to string
monthly_sales.set_index("Month", inplace=True)

#  3 Check for Stationarity (ADF Test)
def adf_test(series):
    result = adfuller(series)
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    return result[1] < 0.05  # If p-value < 0.05, data is stationary

print("\nChecking Stationarity Before Differencing:")
if not adf_test(monthly_sales["Sales"]):
    monthly_sales["Sales"] = monthly_sales["Sales"].diff().dropna()
    print("\nChecking Stationarity After Differencing:")
    adf_test(monthly_sales["Sales"])  # Check stationarity again

#  4 Find Optimal ARIMA Parameters (Using Auto ARIMA)
from pmdarima import auto_arima

auto_model = auto_arima(monthly_sales["Sales"].dropna(), seasonal=False, trace=True, suppress_warnings=True)
p, d, q = auto_model.order
print(f"\nOptimal ARIMA Order: ({p}, {d}, {q})")

#  5 Train ARIMA Model
model = ARIMA(monthly_sales["Sales"].dropna(), order=(p, d, q))
model_fit = model.fit()

#  6 Forecast Future Sales
forecast_period = 12  # Predict next 12 months
future_dates = pd.date_range(start=df["Order Date"].max(), periods=forecast_period, freq="M")

forecast = model_fit.forecast(steps=forecast_period)
forecast_df = pd.DataFrame({"Date": future_dates, "Forecasted Sales": forecast})

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error

#  1 Load & Preprocess Data
df = pd.read_csv("FINAL_DATASET.csv", parse_dates=["Order Date"], dayfirst=True)
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
df.dropna(subset=["Order Date"], inplace=True)

# Aggregate Monthly Sales
df["Month"] = df["Order Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()
monthly_sales["Month"] = monthly_sales["Month"].astype(str)
monthly_sales.set_index("Month", inplace=True)

#  2 Split Data (Train = 90%, Test = 10%)
split_point = int(len(monthly_sales) * 0.9)
train, test = monthly_sales.iloc[:split_point], monthly_sales.iloc[split_point:]

#  Fix Zero Sales Values (Prevents `inf%` in MAPE)
test["Sales"] = test["Sales"].replace(0, 0.01)

#  Fill Missing Values (Prevents `nan%` in MAPE)
test.fillna(method="ffill", inplace=True)

#  3 Train SARIMA Model
auto_model = auto_arima(train["Sales"], seasonal=True, m=12, trace=True, suppress_warnings=True)
p, d, q = auto_model.order
P, D, Q, S = auto_model.seasonal_order

sarima_model = SARIMAX(train["Sales"], order=(p, max(d, 1), q), seasonal_order=(P, max(D, 1), Q, S))
sarima_fit = sarima_model.fit()

# Forecast SARIMA
test["SARIMA Forecast"] = sarima_fit.forecast(steps=len(test))

#  4 Train Holt-Winters Model (Auto-Tuned)
best_mape = float("inf")
best_model = None
best_params = None

for trend in ["add", "mul"]:
    for seasonality in ["add", "mul"]:
        for period in [6, 12]:  # Removed 24 to avoid errors
            try:
                model = ExponentialSmoothing(train["Sales"], trend=trend, seasonal=seasonality, seasonal_periods=period)
                fit = model.fit(use_boxcox=True, use_basinhopping=True)
                forecast = fit.forecast(len(test))

                mape = np.mean(np.abs((test["Sales"] - forecast) / test["Sales"])) * 100

                if mape < best_mape:
                    best_mape = mape
                    best_model = fit
                    best_params = (trend, seasonality, period)
            except Exception as e:
                print(f" Model Failed: Trend={trend}, Seasonality={seasonality}, Period={period} | Error: {e}")

# If a valid Holt-Winters model is found
if best_model:
    print(f" Best Holt-Winters Parameters: Trend={best_params[0]}, Seasonality={best_params[1]}, Seasonal Period={best_params[2]}")
    print(f" New Improved Holt-Winters MAPE: {best_mape:.2f}%")
    test["Holt-Winters Forecast"] = best_model.forecast(len(test))
else:
    print(" No valid Holt-Winters model found!")

#  5 Evaluate Model Accuracy
mae_sarima = mean_absolute_error(test["Sales"], test["SARIMA Forecast"])
rmse_sarima = np.sqrt(mean_squared_error(test["Sales"], test["SARIMA Forecast"]))
mape_sarima = np.mean(np.abs((test["Sales"] - test["SARIMA Forecast"]) / test["Sales"])) * 100

if best_model:
    mae_hw = mean_absolute_error(test["Sales"], test["Holt-Winters Forecast"])
    rmse_hw = np.sqrt(mean_squared_error(test["Sales"], test["Holt-Winters Forecast"]))
    mape_hw = np.mean(np.abs((test["Sales"] - test["Holt-Winters Forecast"]) / test["Sales"])) * 100
else:
    mae_hw, rmse_hw, mape_hw = None, None, None

print("\n Model Accuracy Comparison:")
print(f" SARIMA MAPE: {mape_sarima:.2f}%")
if best_model:
    print(f" Holt-Winters MAPE: {mape_hw:.2f}%")

#  6 Visualization
plt.figure(figsize=(12,6))
plt.plot(test.index, test["Sales"], label="Actual Sales", marker="o", color="blue")
plt.plot(test.index, test["SARIMA Forecast"], linestyle="dashed", color="red", marker="o", label="SARIMA Forecast")
if best_model:
    plt.plot(test.index, test["Holt-Winters Forecast"], linestyle="dashed", color="green", marker="o", label="Holt-Winters Forecast")

plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("SARIMA vs Holt-Winters Sales Forecast")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from pmdarima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error
from scipy.stats import boxcox
from scipy.special import inv_boxcox

#  1 Load & Preprocess Data
df = pd.read_csv("FINAL_DATASET.csv", parse_dates=["Order Date"], dayfirst=True)
df["Order Date"] = pd.to_datetime(df["Order Date"], errors="coerce")
df.dropna(subset=["Order Date"], inplace=True)

# Aggregate Monthly Sales
df["Month"] = df["Order Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Sales"].sum().reset_index()
monthly_sales["Month"] = monthly_sales["Month"].astype(str)
monthly_sales.set_index("Month", inplace=True)

#  2 Optimize Train-Test Split (97% Train, 3% Test)
split_point = int(len(monthly_sales) * 0.97)  # More train data = Better Predictions
train, test = monthly_sales.iloc[:split_point], monthly_sales.iloc[split_point:]

#  3 Apply Box-Cox Transformation (Better than Log Transformation)
train["Sales"], lambda_value = boxcox(train["Sales"] + 1)  # Box-Cox requires positive values

#  4 Find Best SARIMA Order (p, d, q) & (P, D, Q, S)
auto_model = auto_arima(train["Sales"], seasonal=True, m=12, trace=True, suppress_warnings=True)
p, d, q = auto_model.order
P, D, Q, S = auto_model.seasonal_order
print(f" Best SARIMA Order: ({p}, {d}, {q}) x ({P}, {D}, {Q}, {S})")

#  5 Train SARIMA Model
sarima_model = SARIMAX(train["Sales"], order=(p, d, q), seasonal_order=(P, D, Q, S))
sarima_fit = sarima_model.fit()

#  6 Forecast Future Sales (Convert Back from Box-Cox)
test["SARIMA Forecast"] = inv_boxcox(sarima_fit.forecast(steps=len(test)), lambda_value) - 1  # Reverse Box-Cox

#  7 Evaluate Model Accuracy
mae = mean_absolute_error(test["Sales"], test["SARIMA Forecast"])
rmse = np.sqrt(mean_squared_error(test["Sales"], test["SARIMA Forecast"]))
mape = np.mean(np.abs((test["Sales"] - test["SARIMA Forecast"]) / test["Sales"])) * 100

print("\n SARIMA Model Accuracy Metrics (Optimized):")
print(f" Mean Absolute Error (MAE): {mae:.2f}")
print(f" Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f" Mean Absolute Percentage Error (MAPE): {mape:.2f}%")

#   Visualization
plt.figure(figsize=(12,6))
plt.plot(test.index, test["Sales"], label="Actual Sales", marker="o", color="blue")
plt.plot(test.index, test["SARIMA Forecast"], linestyle="dashed", color="red", marker="o", label="SARIMA Forecast")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Optimized SARIMA Sales Forecast (MAPE < 10%)")
plt.legend()
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
