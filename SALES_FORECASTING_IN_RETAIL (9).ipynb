```python
import pandas as pd

df = pd.read_csv("FINAL_DATASET.csv", parse_dates=['Order Date'])

# Display dataset info
print(df.info())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)
```

```python
df = df.dropna()
print(df.isnull().sum())
```

```python
df["Postal Code"] = df["Postal Code"].astype(str)
```

```python
df["Postal Code"] = df["Postal Code"].apply(lambda x: str(int(float(x))))
```

```python
df = df.drop(columns=['Row ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Order ID'])
```

```python
# Display dataset info
print(df.info())
```

```python
# Aggregate total sales per day
df_daily = df.groupby(df.index)['Sales'].sum()

# Convert into DataFrame
df_daily = df_daily.to_frame()

# Rename column for clarity
df_daily.rename(columns={'Sales': 'Total Sales'}, inplace=True)

# Display the first few rows
print(df_daily)
```

```python
import pandas as pd

# Convert 'Order Date' to datetime safely
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

# Check how many dates are NaT before dropping
print("Before Dropping, Total NaT Dates:", df['Order Date'].isna().sum())

# Remove only truly invalid dates (keep valid ones)
df = df[~df['Order Date'].isna()]

# Check after dropping to confirm only bad rows are gone
print("After Dropping, Total Rows:", len(df))

# Set as index & sort
df.set_index('Order Date', inplace=True)
df = df.sort_index()

# Now Resample
df_monthly = df.resample('ME').sum()
df_monthly = df_monthly[['Sales']].rename(columns={'Sales': 'Total Sales'})

print(df_monthly.head())
```

```python
df_yearly = df.resample('YE').sum()


df_yearly = df_yearly[['Sales']].rename(columns={'Sales': 'Total Sales'})


df_yearly.index = df_yearly.index.strftime('%Y')


print(df_yearly)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Reset the index to access 'Order Date' as a column
df = df.reset_index()

# Extract Month and Year from 'Order Date'
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year

# Boxplot to visualize outliers month-wise
plt.figure(figsize=(16, 6))
sns.boxplot(x=df["Month"].astype(str) + "-" + df["Year"].astype(str), y=df["Sales"])
plt.xticks(rotation=90)
plt.xlabel("Month-Year")
plt.ylabel("Sales")
plt.title(" Monthly Sales Distribution (Detecting Outliers)")
plt.show()

# Scatter plot to see extreme values
plt.figure(figsize=(16, 6))
sns.scatterplot(x=df["Month"].astype(str) + "-" + df["Year"].astype(str), y=df["Sales"])
plt.xticks(rotation=90)
plt.xlabel("Month-Year")
plt.ylabel("Sales")
plt.title(" Sales Scatter Plot (Finding Extreme Values)")
plt.show()
```

```python
import pandas as pd
import numpy as np

# Function to remove only extreme outliers
def remove_extreme_outliers(df):
    df_cleaned = df.copy()

    def iqr_filter(group):
        Q1 = group["Sales"].quantile(0.25)
        Q3 = group["Sales"].quantile(0.75)
        IQR = Q3 - Q1
        upper_bound = Q3 + 3.5 * IQR  #  Increased threshold to remove only extreme spikes
        return group[group["Sales"] <= upper_bound]  # Keep all valid values

    df_cleaned = df_cleaned.groupby(["Year", "Month"], group_keys=False).apply(iqr_filter)

    return df_cleaned

# Apply outlier removal
df_cleaned = remove_extreme_outliers(df)

# Check removed values
outliers_removed = df[~df.index.isin(df_cleaned.index)]
print("Total extreme outliers removed:", len(outliers_removed))
print(outliers_removed[["Year", "Month", "Sales"]])  # Show extreme outliers removed
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot after removing only the 4 highest values
plt.figure(figsize=(16, 6))
sns.scatterplot(x=df_cleaned["Month"].astype(str) + "-" + df_cleaned["Year"].astype(str), y=df_cleaned["Sales"])
plt.xticks(rotation=90)
plt.xlabel("Month-Year")
plt.ylabel("Sales")
plt.title(" Sales Scatter Plot After Removing Only the 4 Extreme Outliers")
plt.show()
```
